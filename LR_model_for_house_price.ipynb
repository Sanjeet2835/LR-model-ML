```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0,1.5,4.5, 5.0,3.0])
y_train = np.array([300.0,500.0,400.0, 750.0, 850.0,600.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
#m is the number of training examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0] #m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^{i}, y^{i}):({x_i} {y_i})")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (1000s of dollars)")
```

```python
#here we are taking w,b manually but w,b can be found out using gradient descent algorithm
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    m = len(x_train)
    f_wb = np.zeros(m)
    for i in range(m):
           f_wb[i]=f_wb[i]+w*x_train[i]+b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
print(f"tmp_f_wb = {tmp_f_wb}")
```

```python
plt.plot(x_train, tmp_f_wb, c='b',label='Predicted Values', marker='s')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (1000s of dollars)")
plt.legend()
plt.show()
```
